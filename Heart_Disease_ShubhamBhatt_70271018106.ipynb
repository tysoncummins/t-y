```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.dtypes
```

```python
numerical = df[['age','trestbps','chol','thalach','oldpeak']]
```

```python
numerical
```

```python
col_names = list(numerical)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x=scaler.fit_transform(numerical)
```

```python
numerical=pd.DataFrame(x)
numerical.columns= col_names
```

```python
numerical
```

```python
df.columns
```

```python
categorical = df.drop(['age','trestbps','chol','thalach','oldpeak'],1)
```

```python
categorical
```

```python
df1 = pd.concat([numerical,categorical],1)
```

```python
df1
```

```python
X = df1.drop(['target'],1)
```

```python
X
```

```python
y = df['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm =confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#KNN Classification
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm =confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion="entropy", random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm =confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=10,criterion="entropy", random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm =confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
