```python
X=car.drop(["Selling_Price","Car_Name"],1)
y=car["Selling_Price"]
```

```python
X=pd.get_dummies(X)
```

```python
X
```

```python
col_names=list(X)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(x)
X.columns= col_names
from sklearn.metrics import r2_score,mean_squared_error,median_absolute_error
```

```python
#Knn
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=7)
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
#np.sqrt(mean_squared_error(y_test,y_pred))
r2_score(y_test,y_pred)
```

```python
#Knn using grid search
params = {"n_neighbors":np.arange(1,16)}
knn = KNeighborsRegressor()
knn_cv = GridSearchCV(knn,param_grid=params, cv = 10,scoring="r2")
knn_cv.fit(X_train,y_train)
best_model = knn_cv.best_estimator_
knn_cv.best_score_
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression,RidgeCV,LassoCV,ElasticNetCV
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
y_pred= lin_reg.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
r2_score(y_test,y_pred)
```

```python
#Lasso
lasso = LassoCV(alphas=np.arange(0.1,2,0.1))
lasso.fit(X_train,y_train)
lasso.score(X_test,y_test)
```

```python
#lasso using grid search
from sklearn.linear_model import Lasso
lasso = Lasso()
params = {"alpha":np.arange(0.1,2,0.1)}
lasso_g_cv =GridSearchCV(lasso, param_grid=params, cv = 10)
lasso_g_cv.fit(X_train,y_train)
```

```python
lasso_g_cv.best_score_
```

```python
#Ridge
ridge = RidgeCV(alphas=np.arange(0.1,2,0.1))
ridge.fit(X,y)
ridge.score(X,y)
```

```python
#Ridge through grid search
from sklearn.linear_model import Ridge
ridge = Ridge()
params = {"alpha":np.arange(0.1,2,0.1)}
ridge_g_cv =GridSearchCV(ridge, param_grid=params, cv = 10)
ridge_g_cv.fit(X_train,y_train)
```

```python
ridge_g_cv.best_score_
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
dtr.score(X_test,y_test)
```
