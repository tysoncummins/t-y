```python
def clean_numerical(column):
    mean = df[df[column]!="?"][column].astype(float).mean()
    df[column] = df[column].str.replace("?", str(mean)).astype(float)
    return df
```

```python
def clean_categorical(column):
    mean = df[df[column]!="?"][column].astype(str).mode()
    df[column] = df[column].str.replace("?", str(mean)).astype(str)
    return df
```

```python
X=df.drop(["price","make"],1)
```

```python
y=df["price"]
```

```python
X=pd.get_dummies(X)
```

```python
X
```

```python
X.dtypes
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
```

```python
col_names=list(X)
```

```python
col_names
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x=scaler.fit_transform(X)
```

```python
x
```

```python
X=pd.DataFrame(x)
X.columns= col_names
```

```python
X
```

```python
from sklearn.metrics import r2_score,mean_squared_error,median_absolute_error
```

```python
#lINEAR REGRESSION
from sklearn.linear_model import LinearRegression,RidgeCV,LassoCV,ElasticNetCV
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
y_pred= lin_reg.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
r2_score(y_test,y_pred)
```

```python
#KNN REGRESSOR
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#KNN USING GRID SEARCH
params = {"n_neighbors":np.arange(1,16)}
knn = KNeighborsRegressor()
knn_cv = GridSearchCV(knn,param_grid=params, cv = 10,scoring="r2")
knn_cv.fit(X_train,y_train)
best_model = knn_cv.best_estimator_
knn_cv.best_score_
```

```python
#LASSO 
lasso = LassoCV(alphas=np.arange(0.1,2,0.1))
lasso.fit(X_train,y_train)
lasso.score(X_test,y_test)
```

```python
#LASSO WITH GRID SEARCH
from sklearn.linear_model import Lasso
lasso = Lasso()
params = {"alpha":np.arange(0.1,2,0.1)}
lasso_g_cv =GridSearchCV(lasso, param_grid=params, cv = 10)
lasso_g_cv.fit(X_train,y_train)
```

```python
lasso_g_cv.best_score_
```

```python
#RIDGE
ridge = RidgeCV(alphas=np.arange(0.1,2,0.1))
ridge.fit(X,y)
ridge.score(X,y)
```

```python
#RIDGE WITH GRID SEARCH
from sklearn.linear_model import Ridge
ridge = Ridge()
params = {"alpha":np.arange(0.1,2,0.1)}
ridge_g_cv =GridSearchCV(ridge, param_grid=params, cv = 10)
ridge_g_cv.fit(X_train,y_train)
```

```python
ridge_g_cv.best_score_
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
dtr.score(X_test,y_test)
```
